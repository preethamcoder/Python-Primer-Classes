### Save time when you code. Do things over and over again!

### Print me all the odd numbers from 1 to 100

````python
# 100-2 = 98
for ind in range(1, 100, 2):
    print(ind)
````

### Display all the multiples of 6 between 0 and 120.

````python
# 126-6 = 120
for ind in range(0, 126, 6):
    print(ind)
````

### Display all numbers between 100 and 500.

````python
# 500 = 501 - 1
for num in range(100, 501):
    print(num)
````

### Print me the multiples of 3 from 0 to 30.

````python
# you find the end limit of a for loop by subtracting the ending value and the step counter. So in this case, 31-3 = 28
for ind in range(0, 31, 3):
    print(ind)
````

### How do you tell if a number is divisible by 2?

You check if it is divisble by 2

If the remainder of a number divided by 2 is 0. then it is divisible by 2. Or else, it is not.

We use the ```%``` sign to find the remainder of a division.

````python
44 % 2
````

````python
44/2
````

If you want to check if 2 things are the same, you use ```==``` sign

A single ```=``` sign is used to ```assign``` value to something

````python
45 % 2 == 1
````

### If the remainder of a number when divided by 2 is 0, it is even. If the remainder of a number when divided by 2 is 1, it is odd.

````python
for ind in range(1, 21):
    if ind % 2 == 0:
        print(ind, "is even")
    elif ind % 2 == 1:
        print(ind, "is odd")
````

### For all numbers between 1 and 250, print "STRIKE" if the number is odd, and "BOWL" if the number is even

````python
for ind in range(1, 251):
    if ind % 2 == 0:
        print("BOWL")
    else:
        print("STRIKE")
````
